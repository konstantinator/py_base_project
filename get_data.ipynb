```python
from tqdm import trange
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import multiprocessing as mps
from tqdm.auto import tqdm

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--window-size=1420,1080')
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
driver = webdriver.Chrome(options=chrome_options)
```

```python
dates = [(j.month, j.day) for j in pd.date_range('2022-01-01', '2022-12-31')]
links = [ f'https://nplus1.ru/news/2022/{month:02}/{day:02}' for month, day in dates]
```

```python
 import multiprocessing as mps
from multiprocessing.pool import ThreadPool as Pool

 
 def get_html(link):
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--window-size=1420,1080')
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-gpu')
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(link)
    innerHTML = driver.execute_script("return document.body.innerHTML")
    soup = BeautifulSoup(innerHTML, 'html.parser')
    driver.close()
    res = [i['href'] for i in soup.find_all('a', class_=lambda x: x and x.startswith('n1_climb_4'))]
    return res

with Pool() as p:
    temp = links[:]
    res = p.map(get_html, tqdm(temp, total=len(temp)))
```
